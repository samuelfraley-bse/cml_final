Missing Rate

```python
missing_summary = (
    train_clf.isna()
             .mean()
             .reset_index()
             .rename(columns={'index': 'column', 0: 'missing_rate'})
)

# Convert to percentage
missing_summary['missing_rate_percent'] = missing_summary['missing_rate'] * 100

# Sort by highest missing rate
missing_summary = missing_summary.sort_values('missing_rate', ascending=False)

print(missing_summary)
```

Look at ID columns

```python
id_cols = ['subject_id', 'hadm_id', 'icustay_id']

for c in id_cols:
    n_rows = train_clf.shape[0]
    n_unique = train_clf[c].nunique()
    print(f"{c}: {n_unique} unique values over {n_rows} rows")
```

```python
icu_per_subject = (
    train_clf.groupby('subject_id')['icustay_id']
             .nunique()
             .sort_values(ascending=False)
)

print(icu_per_subject.head(10))          # top patients with most ICU stays
print(icu_per_subject.value_counts())    # how many have 1, 2, 3... stays
```

```python
hadm_per_subject = (
    train_clf.groupby('subject_id')['hadm_id']
             .nunique()
             .sort_values(ascending=False)
)

print(hadm_per_subject.head(10))
print(hadm_per_subject.value_counts())
```

Find what's correlated to target

```python
target = 'HOSPITAL_EXPIRE_FLAG'

num_cols = (
    train_clf.select_dtypes(include=['number'])
             .columns.drop(target)
)

corr_with_target = (
    train_clf[num_cols]
    .corrwith(train_clf[target])
    .sort_values(ascending=False)
)

corr_with_target.head(20)
```
